```python
#AGE VS CHARGES
df2 = df[['age', 'charges']]
df2.head()
```

```python
#AGE VS CHARGES PLOT
sns.lmplot(x ="age", y ="charges", data = df2, order = 2, ci = None)
```

```python
#BMI VS SALARY LINEAR REGRESSION
pd.options.mode.chained_assignment = None  # default='warn'
df2.fillna(method ='ffill', inplace = True)
X = np.array(df2['age']).reshape(-1, 1)
y = np.array(df2['charges']).reshape(-1, 1)
df2.dropna(inplace = True)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25)
Linreg = LinearRegression()
Linreg.fit(X_train, y_train)
print("Accuracy (Score) : ",Linreg.score(X_test, y_test))
```

```python
y_pred = Linreg.predict(X_test)
plt.scatter(X_test, y_test, color ='b')
plt.plot(X_test, y_pred, color ='k')
plt.show()
```

```python
#TAKING JUST 500 ENTRIES OF AGE TO CHECK THE ACCURACY
df2_500 = df2[:][:500]
sns.lmplot(x ='age', y ="charges", data = df2_500,order = 2, ci = None)
```

```python
df2_500.fillna(method ='ffill', inplace = True)
 
X = np.array(df2_500['age']).reshape(-1, 1)
y = np.array(df2_500['charges']).reshape(-1, 1)
 
df2_500.dropna(inplace = True)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25)
 
regr = LinearRegression()
regr.fit(X_train, y_train)
print("Accuracy (Score) : ",regr.score(X_test, y_test))
```

```python
# PREDICTED PLOT
y_pred = regr.predict(X_test)
plt.scatter(X_test, y_test, color ='b')
plt.plot(X_test, y_pred, color ='k')
 
plt.show()
```
