```python
import seaborn as sns
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
#PAIRPLOT OF WORLDOMETER DATA FOR POPULATION,TOTAL CASES AND TOTAL DEATHS
df=pd.read_csv(r"C:\Users\skand\Desktop\Cognizance\AI\Task-2\worldometer_data.csv")
df2=df.head(50)
sns.pairplot(df2,x_vars=['Population','TotalCases','TotalDeaths'],y_vars=['Population','TotalCases','TotalDeaths'],height=5)
```
