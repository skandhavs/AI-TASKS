```python
import seaborn as sns
import pandas as pd
```

```python
#REFERENCE CELL FOR ADANI PORTS
df=pd.read_csv(r"C:\Users\skand\Desktop\Cognizance\AI\Task-2\archive\ADANIPORTS.csv")
df.head()              
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#I have just plotted for 25 days and their turnover
df2=df.head(25)
fig,ax = plt.subplots(figsize=(35,20))
plt.ylabel('Turnover', fontsize=30)
plt.xticks(fontsize=13)
plt.yticks(fontsize=20)
plt.title('ADANI PORTS-BAR PLOT',fontsize=45 )
plt.xlabel('Date',fontsize=30)
sns.barplot(x=df2['Date'],y=df2['Turnover'],ax=ax)
```

```python
fig,ax = plt.subplots(figsize=(10,10)) 
plt.title('ADANI PORTS-HIST PLOT',fontsize=20 )
plt.xlabel('VWAP',fontsize=20)
sns.histplot(x=df['VWAP'],kde=True,ax=ax)
```

```python
fig,ax = plt.subplots(figsize=(15,10)) 
plt.title('ADANI PORTS-SCATTER PLOT',fontsize=20 )
sns.scatterplot(x=df['Open'],y=df['Close'],ax=ax)
```

```python
fig,ax = plt.subplots(figsize=(15,15)) 
plt.title('ADANI PORTS-HEAT MAP',fontsize=20 )
sns.heatmap(df2.corr(),annot=True,ax=ax)
#NaN Values are white in colour
```
