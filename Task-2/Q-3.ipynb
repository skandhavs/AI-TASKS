```python
import matplotlib.pyplot as plt
import random
import numpy as np
from scipy.interpolate import make_interp_spline
amplitude = 10
t = 10
random.seed()
#FIRST RANDOM FUNCTION/SIGNAL
rand = np.random.uniform(3, 4)
rand2 = np.arange(t)
x1= amplitude * np.cos(-rand2/rand)
y1=np.tan(x1)
plt.subplot(2,1,1)
plt.plot(x1,y1)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Random function 1")
plt.show()
#SECOND RANDOM FUNCTION/SIGNAL
random.seed()
rand= np.random.uniform(5, 9)
rand2 = np.arange(t)
x2=amplitude * np.sin(rand2/rand)
y2=np.tan(x2)
X_Y_Spline = make_interp_spline(x2, y2)
X2 = np.linspace(x2.min(), x2.max(), 500)
Y2 = X_Y_Spline(X2)
plt.subplot(2,1,2)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Random function 2")
plt.plot(X2,Y2)
plt.show()

#ADDING TWO SIGNALS
x3=x1
#x3=x1,since domain of x1 is larger and covers x2 too
y3=np.tan(x1)+np.tan(x2)
plt.plot(x3,y3)
plt.title("Addition of two the two functions")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```
