```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
from sklearn.datasets import load_digits
```

```python
mnist=load_digits()
```

```python
type(mnist)
```

```python
df = pd.DataFrame(data= np.c_[mnist['data'], mnist['target']],columns= mnist['feature_names'] + ['target'])
```

```python
df.head()
```

```python
y=df["target"]
y.head()
```

```python
df.drop("target",axis=1,inplace=True)
df.head()
```

```python
x_tr,x_te,y_tr,y_te=train_test_split(df,y,test_size=0.30,random_state=0)
classif=RandomForestClassifier(n_estimators = 16,criterion='entropy',random_state=0)
classif.fit(x_tr,y_tr)
```

```python
y_pr=classif.predict(x_te)
y_pr
```

```python
cm=confusion_matrix(y_te,y_pr)
print("Cofusion MAtrix:",cm)
```

```python
print("Accuracy Score:",accuracy_score(y_te,y_pr))
```

```python
print(classification_report(y_te,y_pr))
```
